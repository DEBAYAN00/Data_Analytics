```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sb
```

```python
weather = pd.read_csv('fiveyearsweatherdata.csv')X = weather['Temperature']
y = weather['Relative Humidity']
```

```python
weather.head()
```

```python
weather.info()
```

```python
weather.describe()
```

```python
ax = np.array(X).reshape(-1,1)
by = np.array(y)
```

```python
ax_train,ax_test,by_train,by_test = train_test_split(ax,by,test_size=0.2,random_state=0)
```

```python
mod = LinearRegression()
mod.fit(ax_train,by_train)
```

```python
prediction = mod.predict([[40]])
prediction
```

```python
y_pred = mod.predict(ax_test)
intercept = mod.intercept_
intercept
#By = y_pred.reshape(-1,1)
#By
```

```python
mod.coef_
```

```python
#Scatter plot betwn. 'Acctual Relative Humidity'vs 'Predicted Relative Humidity'

plt.plot(by_test,y_pred,'o')
plt.xlim(40,100)
plt.ylim(40,100)
plt.show()
```

```python
#Scatter plot between 'Temparature(Test)' Vs 'Relative Humidity(Test)'
plt.plot(ax_test,by_test,'o')
plt.xlim(0,100)
plt.ylim(0,100)
plt.show()
```

```python
#Scatter plot between 'Temparature(Train)' Vs 'Relative Humidity(Train)'
plt.plot(ax_train,by_train,'o')
plt.xlim(0,100)
plt.ylim(0,100)
plt.show()
```
