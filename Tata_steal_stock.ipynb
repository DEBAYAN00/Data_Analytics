```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Tata-steel.csv')
```

```python
X = np.array(df['Date']).reshape(-1,1)
y = np.array(df['Low Price']).reshape(-1,1)
```

```python
mod1 = LinearRegression()
mod1.fit(X,y)
```

```python
pred_low = mod1.predict(X)
pred_low
```

```python
X = np.array(df['Date']).reshape(-1,1)
y = np.array(df['High Price']).reshape(-1,1)
```

```python
mod2 = LinearRegression()
mod2.fit(X,y)
```

```python
pred_high = mod2.predict(X)
pred_high
```

```python
#Best fit Regression line for High Price

plt.plot(df['Date'],pred_high)
plt.plot(df['Date'],df['High Price'],color='g')
plt.show()
```

```python
plt.plot(df['Date'],df['Low Price'],color='orange')
plt.plot(df['Date'],pred_low,color='b')
plt.show()
```

```python
plt.plot(X,df['Low Price'],color = 'red')
plt.plot(X,df['High Price'],color = 'green')
plt.show()
```
